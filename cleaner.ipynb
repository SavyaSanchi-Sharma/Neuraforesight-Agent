```python
import pandas as pd
from rdkit import Chem
```

```python
drugBankVocab=pd.read_csv("data/drugbank vocabulary.csv")
drugBankVocab
```

```python
drugBankVocab.columns
```

```python
supplier = Chem.SDMolSupplier("data/open structures.sdf")
drugBankStructuredata = []
for mol in supplier:
    if mol is None:
        continue   
    props = mol.GetPropsAsDict()
    props = {k: str(v) for k, v in props.items()}  
    props['SMILES'] = Chem.MolToSmiles(mol)
    props['Name'] = mol.GetProp('_Name') if mol.HasProp('_Name') else ''
    drugBankStructuredata.append(props)
drugBankStructuredata = pd.DataFrame(drugBankStructuredata)
```
